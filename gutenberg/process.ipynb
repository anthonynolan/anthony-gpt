```python
from gutenberg_clean import read_clean_gutenberg_book

with open('english-books.txt') as f:
    content = f.readlines()


book_count = 150
a = [read_clean_gutenberg_book('./'+'/'.join(f.strip().split('/')[1:])) for f in content[:book_count]]

big_book_chunk = ' '.join(a)
print(f'approx word count: {len(big_book_chunk.split()):,}')
```

```python
with open('../data/big_book_chunk.txt', 'wt') as f:
    f.write(big_book_chunk)
```
