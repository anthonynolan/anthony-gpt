```python
import numpy as np
import matplotlib.pyplot as plt
import random

random.seed(455)
x = np.arange(1000)
x = np.expand_dims(x, -1)
y = np.array([random.random()*500 +val for val in x*2+4 ])
y = np.expand_dims(y, -1)

relu = lambda i: np.maximum(0,i) # noqa: E731

batch_size = 10
units_1 = 3

W1 = np.random.randn(x.shape[1], units_1)
b1 = np.zeros([y.shape[1], units_1])

W_out = np.random.randn(units_1, 1)
b_out = np.zeros((1,1))

first_batch_outputs = relu((x[0:batch_size]@W1+b1)@W_out+b_out)

plt.plot(first_batch_outputs)
plt.plot(y[:batch_size].reshape(10,1))
```
