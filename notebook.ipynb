```python
import numpy as np
```

```python
row_length = 6
```

```python
with open('persuasion.txt') as f:
    content = f.read()
words = content.split()
len(words)
```

```python
blocks = []
for i in range(0, len(words), row_length):
    blocks.append(words[i:i+row_length])
print(blocks[:5])
```

```python
training_data = np.array(blocks[:-1]).reshape(-1, row_length)
print(training_data)
```
